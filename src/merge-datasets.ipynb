```python
merged = pd.merge(grouped_cars, grouped_reviews, how='inner', on = 'dealer_id', suffixes=('_cars', '_rev'))
```

```python
merged = merged[['dealer_id', 'count_cars', 'count_rev']]
```

```python
merged
```

```python
merged['diff'] = merged['count_rev'] - merged['count_cars']
```

```python
merged
```

```python
#filter reviews so that we either take all reviews if there are more cars, or we take as many reviews
#as nr of cars (if there are more reviews than cars)
r = []
for i in range(len(merged)):
    dealer = merged['dealer_id'].loc[i]
    nr_cars = merged['count_cars'].loc[i]
    rev = reviews[reviews['dealer_id'] == dealer]
    if len(rev) == 0: 
        continue
    if merged['diff'].loc[i] < 0 :
        r.append(rev)
    else:
        r.append(rev.sample(n=int(nr_cars)))
        
r = pd.concat(r)                 
```

```python
r
```

```python
#keep only the cars that are in the new review dataset
cars = cars[cars['dealer_id'].isin(r['dealer_id'])]
```

```python
cars = cars.reset_index()
```
